# Прогнозирование уровня средней заработной платы в России

В проекте используюся о заработной плате за каждый месяц с января 1993 по август 2016. Если хотите, можете дописать в конец ряда данные за следующие месяцы, если они уже опубликованы; найти эти данные можно, например, здесь.

Необходимо проанализировать данные, подобрать для них оптимальную прогнозирующую модель в классе ARIMA и построить прогноз на каждый месяц на два года вперёд от конца данных.

```python
%pylab inline
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
from itertools import product
```

### Загрузка данных:

```python
data = pd.read_csv('WAG_C_M.csv', sep = ';', index_col = ['month'], parse_dates = ['month'], dayfirst = True)
```

```python
data.rename(columns = {'WAG_C_M':'wage'}, inplace = True)
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

### График:

```python
plt.figure(figsize(15,5))
data.wage.plot();
plt.ylabel('Average Wage')
plt.xlabel('Month')
plt.title('Average wage')
```

### Стационарность и STL-декомпозиция

```python
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(data.wage).plot();
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data.wage)[1])
```

**Гипотеза о нестационарности ряда не отвергается.**

**Особенности временного ряда:** восходящий тренд, годовая сезонность

### Стабилизация дисперсии

Для стабилизации дисперсии проведем преобразование Бокса-Кокса:

```python
data['wage_boxcox'], lmbda = stats.boxcox(data.wage)
```

```python
plt.figure(figsize(15,5))
data.wage_boxcox.plot()
plt.xlabel('Month')
plt.ylabel('Average Wage (transformed)')
```

```python
print("Опитимум параметра Бокса-Кокса: %f" % lmbda)
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data.wage_boxcox)[1])
```

### Стационарность

Попробуем сезонное дифференцирование; сделаем на продифференцированном ряде STL-декомпозицию и проверим стационарность:

```python
data['wage_boxcox_diff'] = data.wage_boxcox - data.wage_boxcox.shift(12)
```

```python
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(data.wage_boxcox_diff[12:]).plot();
```

```python
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data.wage_boxcox_diff[12:])[1])
```

По критерию Дики-Фуллера можно отвергнеть гипотезу о нестационарности ряда (0.0147 < 0.05). Присутствует тренд, попробуем избавиться от него, применив еще одно дифференцирование:

```python
data.columns
```

```python
data['wage_boxcox_diff2'] = data['wage_boxcox_diff'] - data['wage_boxcox_diff'].shift(1)
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(data['wage_boxcox_diff2'][13:]).plot()   
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data['wage_boxcox_diff2'][13:])[1])
```

Гипотеза о нестационарности отвергается, тренд отсутствует.

## Подбор модели

ACF и PACF полученного ряда:

```python
plt.figure(figsize(15,10))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(data.wage_boxcox_diff[12:].values.squeeze(), lags=48, ax=ax)
pylab.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(data.wage_boxcox_diff[12:].values.squeeze(), lags=48, ax=ax)
pylab.show()
```

**Параметры модели:** Q = 0, q = 1, P = 1, p = 1

```python
ps = range(0, 2)
d = 1
qs = range(0, 2)
Ps = range(0, 2)
D = 1
Qs = range(0, 1)
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
parameters_list
```

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(data.wage_boxcox, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

### Лучшая модель:

```python
print(best_model.summary())
```

### Остатки:

```python
plt.figure(figsize(15, 10))
plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel('Residuals')
plt.xlabel('Month')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[13:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

* Критерией Стьюдента: p = 0.09 > 0.05 - __остатки несмещены __
* Критерий Дики-Фуллера: p < 0.05 - __остатки стационарны__

### Увеличим диапазон параметров:

```python
ps = range(0, 4)
d = 1
qs = range(0, 4)
Ps = range(0, 2)
D = 1
Qs = range(0, 2)
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
```

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(data.wage_boxcox, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

* Критерией Стьюдента: p = 0.276 > 0.05 - __остатки несмещены __
* Критерий Дики-Фуллера: p < 0.05 - __остатки стационарны__

```python
# Функция для расчета обратного преобразвания Бокса-Кокса
def invboxcox(y,lmbda):
   if lmbda == 0:
      return(np.exp(y))
   else:
      return(np.exp(np.log(lmbda*y+1)/lmbda))
```

```python
data['model'] = invboxcox(best_model.fittedvalues, lmbda)
plt.figure(figsize(15,7))
data.wage.plot(label = 'Fact')
data.model[13:].plot(color='r', label = 'Forecast', linestyle = '--')
plt.ylabel('Average wage')
plt.xlabel('Month')
plt.legend()
pylab.show()
```
